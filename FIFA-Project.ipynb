```python
fifa=data_copy[['Name','Age','OVA','Nationality','Club', 'BOV','BP','Position','Value','Wage','Release Clause','Contract']]
fifa
```

```python
fifa.isnull().sum()
```

```python
fifa['Value'] = fifa['Value'].replace({'€': '', ' ': '', 'M': 'e+06', 'K': 'e+03'}, regex=True).astype(float).astype(int)
fifa['Value'] 
```

```python
fifa['Wage'] = fifa['Wage'].replace({'€': '', ' ': '', 'M': 'e+06', 'K': 'e+03'}, regex=True).astype(float).astype(int)
fifa['Wage']
```

```python
fifa['Release Clause'] = fifa['Release Clause'].replace({'€': '', ' ': '', 'M': 'e+06', 'K': 'e+03'}, regex=True).astype(float).astype(int)
fifa['Release Clause']
```

```python
fifa['Release Clause']
```

```python
fifa['Value']=fifa['Value'].astype(int)
fifa['Wage']=fifa['Wage'].astype(int)
fifa['Release Clause']=fifa['Release Clause'].astype(int)
```

```python
fifa.dtypes
```

```python
fifa=fifa.loc[(fifa['Wage']!=0)]
fifa=fifa.loc[(fifa['Age']<=41)] #eliminamos outliers en la edad

#en este paso nos quedamos con los jugadores que tienen salario mayor que 0 (había posibles retirados o situación incierta)
```

```python
fifa.loc[(fifa['Release Clause']<40000000) & (fifa['Age']<=25) & (fifa['BP']=='CB')].sort_values('OVA', ascending=False)[:10]

#PRIMERA PREGUNTA:
#Suponemos que tenemos 40M€ ya que se nos ha lesionado nuestro CB (estamos en mercado de invierno), no podemos negociar un fichaje y debemos pagar la cláusula de alguno.
#Queremos fichar un crack "CB" (Central) con proyección a largo plazo, por eso buscamos a una persona de menos de 25 años porque en la curva 
#vemos que a partir de los 25 hay una reducción visible del margen de mejora y no queremos que se lesione
#Hemos descubierto a través de las anteriores gráficas analizando Age y OVA que a menor edad y a mayor OVA mejor



```

```python
#SEGUNDA PREGUNTA:
#Queremos priorizar la proyección de media utilizando para ello si desputan dentro de su rango edad con calidad /edad

fifa["OVA/Age"]=fifa['OVA']/fifa['Age']

fifa.loc[(fifa['Release Clause']<40000000) & (fifa['Age']<=25) & (fifa['BP']=='CB')].sort_values('OVA/Age', ascending=False)[:10]
```
